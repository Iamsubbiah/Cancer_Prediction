```python
# divide into what to be x(dependent variable)= found (diagonosis1 means cancer and 0 means no)  and y=rest of data used to predict diagonosis(independent variable)

# y=df.diagnosis
df.diagnosis=[ 1 if value == "M" else 0 for value in df.diagnosis]
df
```

```python
# df.diagnosis=============df['diagnosis']

df.diagnosis=df['diagnosis'].astype('category',copy=False)
df.diagnosis.value_counts()
```

```python
df.diagnosis.value_counts().plot(kind='bar')
```

```python
y=df['diagnosis']
print(df)
x=df.drop(['diagnosis'],axis=1)
# axis=0=row and 1 is column 
```

## NORMALIZE the DATA(equal weitage)

```python
# now look we have multiple variables in x so each has different units some like 0.123 and some others like 120 
# if we simply add all and check then priority that it gives to columns that scale is in decimal holds low weitage so only column with high scale do prominance 
# to rectify it we have to normalise and give all columns equal weitage for that we are gonna put all in same scale or weitage

from sklearn.preprocessing import StandardScaler

# creating scaler object

scaler=StandardScaler()
```

```python
# now values to be fit with scaler and transform it making it 2d ready to do logistic training
xscaled=scaler.fit_transform(x)
xscaled
# y no needs to be scaled as it has only one typed column value
```

## SPLIT TRAIN AND TEST DATA

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train ,y_test= train_test_split(xscaled,y,test_size=0.3,random_state=6)
```

## TRAIN THE MODEL

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg=LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
# see now we are finding(PREDICTING Y) with x testing data if it equal to answer of x test prediction answer thats y test then our training is perfect
# once we get predicted y if its equal to y test then mass ji
```
